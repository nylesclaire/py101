Question 2

```python
str1 = "Come over here!"  # True
str2 = "What's up, Doc?"  # False

print(str1[-1] == '!')
print(str2[-1] == '!')
```

```python
def EndsExclamation(string):
    if string[-1] == "!":
        return True
    else:
        return False

str1 = "Come over here!"  # True
str2 = "What's up, Doc?"  # False

print(EndsExclamation(str1))
print(EndsExclamation(str2))
```
