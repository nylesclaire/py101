Question 2

```python
# My solution
def factors(number):
    if number < 0:
        return ("That's a negative number-- try a positive one.")
    divisor = number
    result = []
    while divisor != 0:
        if number % divisor == 0:
            result.append(number // divisor)
        divisor -= 1
    return result

print(factors(42))
print(factors(-42))
```

```python
# Their solution

def factors(number):
    divisor = number
    result = []
    while divisor > 0:
        if number % divisor == 0:
            result.append(number // divisor)
        divisor -= 1
    return result

print(factors(42))
print(factors(-42))
```
